```python
a = []

for j in range(120):
    items = right_tables[0].find_all('td')[j].text
    a.append(items)
    
for j in range(135):
    items = right_tables[1].find_all('td')[j].text
    a.append(items)
    
for j in range(105):
    items = right_tables[2].find_all('td')[j].text
    a.append(items)
    
for j in range(155):
    items = right_tables[3].find_all('td')[j].text
    a.append(items)
    
for j in range(130):
    items = right_tables[4].find_all('td')[j].text
    a.append(items)
```

```python
a[:20]
```

```python
a[::5]
```

```python
actors = a[::5]
character = a[1::5]
season = a[2::5]
episode = a[3::5]
title = a[4::5]
```

```python
title[:4]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['Actors'] = actors
df['Character'] = character
df['Season'] = season
df['Episode'] = episode
df['Title'] = title
```
